#### Imported necessary libraries

```python
from bs4 import BeautifulSoup
import requests
```

```python
# creating variable to store address of the website
url='https://www.w3schools.com/python/python_ref_string.asp'
```

```python
#This function takes a URL as its first argument and returns a response object. 
#The Response object contains all the information that was returned by the server in response to your request.

page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text) # allows us to retrieve human-readable text content from HTML tags
```

```python
print(soup)
```

```python
# for better representation or understanding
print(soup.prettify())
```

```python
table=soup.find_all('table')[0] # here indexing is done to remove list
```

```python
table
```

```python
headers=table.find_all('th') # taking out title of the table
```

```python
headers
```

```python
titles=[h.text.strip() for h in headers] # removing unnecessary tags
```

```python
titles
```

#### importing pandas

```python
import pandas as pd
```

##### assinging title to dataframe

```python
df=pd.DataFrame(columns=titles)
```

```python
df
```

```python
print(len(df))
```

```python
column_data=table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]: # first element is not of use hence started from index=1
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    individual_row_data[1]=individual_row_data[1].replace("\n    ","") # removing additional "\n   " from 2nd element
    print(individual_row_data)
    length=len(df)
    print(length)
    df.loc[length]=individual_row_data
```

```python
df
```

##### deleting duplicate values if exists

```python
df.drop_duplicates(subset=None, keep='first', inplace=True) # for removing duplicates and keeping first values
df
```
